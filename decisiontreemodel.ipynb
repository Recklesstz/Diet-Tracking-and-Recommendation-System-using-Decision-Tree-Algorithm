```python
import pandas as pd
import numpy as np
```

```python
from chefboost import Chefboost as chef
from chefboost.training import Training 
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pickle
```

```python
dataset = pd.read_csv('dietdataset.csv')
```

```python
dataset = pd.DataFrame(data=dataset.iloc[:,0:10].values,columns = ['meal_name','carb','meat','vege','fruit', 'type','vegan','allergy','time'])
le = LabelEncoder()
dataset_encoded = dataset.iloc[:,0:10]
for i in dataset_encoded:
    dataset_encoded[i] = le.fit_transform(dataset_encoded[i])
dataset_encoded
```

```python
dataset
```

```python
X = dataset_encoded.iloc[:,1:10]

y = dataset_encoded.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = DecisionTreeClassifier(criterion="gini")

model.fit(X_train.values, y_train)

```

```python
import random



def input_encode(entry,room): 
    meal = dataset.values.tolist()
    meal_encode = dataset_encoded.values.tolist()
    lists = []
    encode = []   
    for i in entry:
        for j in meal:
            if i==j[room]:
                lists.append(j)
                break
                 
    for j in lists: 
        encode.append(meal_encode[meal.index(j)][room])
        
    return encode
#         return meal_encode[meal.index(j)][room]

def input_decode(entry,room): 
    meal = dataset.values.tolist()
    meal_encode = dataset_encoded.values.tolist()
    lists = []
    decode = []   
    for i in entry:
        for j in meal_encode:
            if i==j[room]:
                lists.append(j)
                break
                 
    for j in lists: 
        decode.append(meal[meal_encode.index(j)][room])
        
    return decode

carb_input = ['None']
meat_input = ['None']
vege_input = ['carrot']
fruit_input = ['None']
type_input = ['stir-fried']
vegan_input = ['yes']
allergy_input = ['None']
time_input = ['Dinner']

carb_encode = input_encode(carb_input,1)[0]
meat_encode = input_encode(meat_input,2)[0]
vege_encode = input_encode(vege_input,3)[0]
fruit_encode = input_encode(fruit_input,4)[0]
type_encode = input_encode(type_input,5)[0]
vegan_encode = input_encode(vegan_input,6)[0]
allergy_encode = input_encode(allergy_input,7)[0]
time_encode = input_encode(time_input,8)[0]

output = []

        
datainput = [carb_encode,meat_encode,vege_encode,fruit_encode,type_encode,vegan_encode,allergy_encode,time_encode]

result = model.predict([datainput])
    
prediction = input_decode(result,0)
    
print("Prediction: "+prediction[0])  
print(carb_input,meat_input,vege_input,fruit_input,type_input,vegan_input,allergy_input,time_input)
```

```python
accuracy_score(model.predict(X_test.values),y_test)
```

```python
average_precision_score(y_test,model.predict(X_test.values),average='macro',pos_label=2, sample_weight=None)
```

```python
f1_score(y_test, model.predict(X_test.values),average='weighted')
```

```python
recall_score(y_test, model.predict(X_test.values),average='weighted')
```

```python
y_true = y_test
y_pred = model.predict(X_test.values)
classification_report(y_true, y_pred, labels=None, target_names=None, sample_weight=None, digits=2, output_dict=True, zero_division=0)
```
